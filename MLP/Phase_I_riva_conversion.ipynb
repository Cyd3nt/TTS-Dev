```python
from nemo.collections.tts.models import FastPitchModel
import torch
```

## Applejack

```python
#Copy the ckpt file
!cp /content/drive/MyDrive/TTS_finetuning/apple_jack/FastPitch--v_loss=1.1907-epoch=132-last.ckpt .
checkpoint = '/content/FastPitch--v_loss=1.1907-epoch=132-last.ckpt'
!mv /content/FastPitch--v_loss=1.1907-epoch=132-last.ckpt fp_applejack.ckpt

#Convert the ckpt file to nemo format
model_checkpoint = '/content/fp_applejack.ckpt'
model =  FastPitchModel.load_from_checkpoint(model_checkpoint)
checkpoint = torch.load(model_checkpoint) 
model.load_state_dict(checkpoint['state_dict'], strict=True) 
model.save_to('applejack_FP.nemo')

#Convert the nemo file to riva format
!nemo2riva --out applejack_fastpitch.riva --key 'convaidev' applejack_FP.nemo

#upload the riva file to the gcp bucket
!gsutil cp /content/applejack_fastpitch.riva gs://tts-voices-mlp/
```
